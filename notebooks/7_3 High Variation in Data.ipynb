## Problem 2 : High Variation in Data - Different size Images

### Solution : Reshaping the images

```python
# Importing the required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading and plotting two random images with different sizes

img_1 = plt.imread('./images/' + 'emergency_100.jpg')
img_2 = plt.imread('./images/' + 'emergency_1000.jpg')

plt.imshow(img_1)
plt.show()
print(img_1.shape)
```

```python
plt.imshow(img_2)
plt.show()
print(img_2.shape)
```

```python
# Importing the load_img module 

from keras.preprocessing.image import load_img, img_to_array
```

```python
img_1 = load_img('./images/' + 'emergency_100.jpg', target_size = (224,224))
img_2 = load_img('./images/' + 'emergency_1000.jpg', target_size = (224,224))

plt.imshow(img_1)
plt.show()
print(img_to_array(img_1).shape)
```

```python
plt.imshow(img_2)
plt.show()
print(img_to_array(img_2).shape)
```
