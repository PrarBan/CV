# DropOut

```python
# Importing the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from keras.layers import Dense
from keras.models import Sequential
from keras.preprocessing import image

from sklearn.model_selection import train_test_split
```

```python
# Loading the Emergency Classification dataset

data = pd.read_csv('../datasets/emergency_classification/emergency_classification.csv')
```

```python
# Seeing the first 5 rows of the dataframe

data.head()
```

```python
# reading all the images in the dataset and converting them into array

images = []
for img_loc in data.image_names:
    img = image.load_img('../datasets/emergency_classification/images/' + img_loc, target_size=(100, 100))
    img = image.img_to_array(img)
    images.append(img)
    
images=np.array(images)
```

```python
# Reshaping 3D images to 1D

train_x = images.reshape(2352, 100*100*3)
```

```python
# Rescaling the images

train_x /= train_x.max()
train_y = data.emergency_or_not.values
```

```python
# splitting the dataset into train and validation

X_train, X_valid, Y_train, Y_valid=train_test_split(train_x,train_y,test_size=0.3, random_state=42)
```

```python
# Overfitted model

model=Sequential()

model.add(Dense(100, input_dim=100*100*3, activation='sigmoid'))
model.add(Dense(50, activation = 'sigmoid'))
model.add(Dense(units=1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer="sgd", metrics=['accuracy'])
```

```python
# fitting the model

history = model.fit(X_train, Y_train, epochs=500, batch_size=128,validation_data=(X_valid,Y_valid))
```

```python
# Plotting training accuracy vs validation accuracy

plt.plot(history.history['acc'], 'r')
plt.plot(history.history['val_acc'], 'b')
plt.legend({'training accuracy': 'r', 'validation accuracy':'b'})
plt.show()
```

```python
# plotting training loss vs validation loss

plt.plot(history.history['loss'], 'r')
plt.plot(history.history['val_loss'], 'b')
plt.legend({'training loss': 'r', 'validation loss':'b'})
plt.show()
```

```python
from keras.layers import Dropout
```

```python
# Adding dropouts to the layers

model=Sequential()

model.add(Dense(100, input_dim=100*100*3, activation='sigmoid'))
model.add(Dropout(0.4))
model.add(Dense(50, activation = 'sigmoid'))
model.add(Dropout(0.4))
model.add(Dense(units=1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer="sgd", metrics=['accuracy'])
```

```python
# fitting the model

history = model.fit(X_train, Y_train, epochs=500, batch_size=128,validation_data=(X_valid,Y_valid))
```

```python
# plotting Training accuracy vs validation accuracy

plt.plot(history.history['acc'], 'r')
plt.plot(history.history['val_acc'], 'b')
plt.legend({'training accuracy': 'r', 'validation accuracy':'b'})
plt.show()
```

```python
# plotting training loss vs validation loss

plt.plot(history.history['loss'], 'r')
plt.plot(history.history['val_loss'], 'b')
plt.legend({'training loss': 'r', 'validation loss':'b'})
plt.show()
```

# Early Stopping

```python
model=Sequential()

model.add(Dense(100, input_dim=100*100*3, activation='sigmoid'))
model.add(Dropout(0.4))
model.add(Dense(50, activation = 'sigmoid'))
model.add(Dropout(0.4))
model.add(Dense(units=1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer="sgd", metrics=['accuracy'])
```

```python
from keras.callbacks import EarlyStopping
```

```python
estop = EarlyStopping(monitor='val_acc', patience=30)
```

```python
# fitting the model

history = model.fit(X_train, Y_train, epochs=1000, batch_size=128,validation_data=(X_valid,Y_valid), callbacks=[estop])
```

```python
# plotting Training accuracy vs validation accuracy

plt.plot(history.history['acc'], 'r')
plt.plot(history.history['val_acc'], 'b')
plt.legend({'training accuracy': 'r', 'validation accuracy':'b'})
plt.show()
```

```python
# plotting training loss vs validation loss

plt.plot(history.history['loss'], 'r')
plt.plot(history.history['val_loss'], 'b')
plt.legend({'training loss': 'r', 'validation loss':'b'})
plt.show()
```
