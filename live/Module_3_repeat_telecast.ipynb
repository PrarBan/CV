```python
# create an empty python list
X = []

# go through all the image locations one by one
for img_name in data.image_names:
    # read the image from location
    img = plt.imread('../datasets/emergency_classification/images/' + img_name)
    # pile it one over the other
    X.append(img)
    
# convert this python list to a single numpy array
X = np.array(X)
```

```python
X.shape
```

```python
y = data.emergency_or_not.values
y.shape
```

```python
from keras.utils import to_categorical

y = to_categorical(y)
y.shape
```

```python
X
```

```python
y
```

```python
X.min(), X.max()
```

```python
from keras.applications.vgg16 import preprocess_input
```

```python
X = preprocess_input(X, mode='tf')
```

```python
X.min(), X.max()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
base_model = VGG16(weights='imagenet', input_shape=(224, 224, 3), include_top=False)
```

```python
base_model.summary()
```

```python
base_model_features = base_model.predict(X_train)
base_model_features_valid = base_model.predict(X_valid)
```

```python
base_model_features.shape
```

```python
base_model_features = base_model_features.reshape(-1, 7*7*512)
base_model_features_valid = base_model_features_valid.reshape(-1, 7*7*512)
base_model_features.shape
```

```python
base_model_features.max()
```

```python
max_val = base_model_features.max()
base_model_features = base_model_features / max_val
base_model_features_valid = base_model_features_valid / max_val
```

```python
base_model_features.min(), base_model_features.max()
```

```python
from keras.models import Sequential
from keras.layers import Dense, InputLayer
```

```python
model = Sequential()
model.add(InputLayer((25088, )))
model.add(Dense(units=1024, activation='relu')) # activation relu, sigmoid, softmax
model.add(Dense(units=2, activation='softmax')) 
```

```python
model.compile(optimizer='sgd', metrics=['accuracy'], loss='categorical_crossentropy')
```

```python
model.summary()
```

```python
model.fit(base_model_features, y_train, epochs=10, validation_data=(base_model_features_valid, y_valid))
```

```python
predictions = model.predict_classes(base_model_features_valid)
predictions
```

```python
prediction_probabilities = model.predict(base_model_features_valid)
prediction_probabilities
```

```python
# pull out the original images from the data which correspond to the validation data
_, valid_vehicles, _, _ = train_test_split(data.image_names.values, y, test_size=0.3, random_state=42)

# get a random image from validation data
img_name = rng.choice(valid_vehicles)

# get the corresponding probability and actual class
prob = (prediction_probabilities[:, 1] * 100).astype(int)[np.where(valid_vehicles == img_name)[0][0]]
actual_class = data.loc[data.image_names == img_name, 'emergency_or_not'].values[0]

# read the image
img = plt.imread('../datasets/emergency_classification/images/' + img_name)

# print probability and actual class
print('There is', prob, "% chance that it is an emergency vehicle")
print('Whereas actual class is', actual_class)

# plot the image
plt.imshow(img)
```
